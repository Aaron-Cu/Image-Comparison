## Methods

### MeanSquareError()
#### Inputs
Two images, imageOne and imageTwo. Must be of the same size.

#### Outputs
A float, the calculated Mean Square error

#### Summary
The Mean Square Error is the average squared differents between two models. The larger the magnitude of the value, the less similar models are.

### imageComparison()
#### Inputs
Two images, imageOne and imageTwo. Must be of the same size. 

#### Summary
Calculates and compares the Mean Square Error to the [Strucural Similarity Index](http://www.cns.nyu.edu/pub/eero/wang03-reprint.pdf).


```python
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
import numpy as np
import cv2

def MeanSquareError (imageOne, imageTwo):
    error = np.sum(imageOne.astype("float") - imageTwo.astype("float"))
    error /= float(imageOne.shape[0] * imageOne.shape[1])
    return error

def imageComparison (imageOne, imageTwo):
    meanSquareError = MeanSquareError(imageOne, imageTwo)
    structuralSimilarityIndex = ssim(imageOne, imageTwo)
    
    figure = plt.figure("Image Comparison")
    plt.suptitle("Mean Square Error: %.2f, Structural Similarity Index Method: %.2f" % (meanSquareError, structuralSimilarityIndex))
    
    ax = figure.add_subplot(1, 2, 1)
    plt.imshow(imageOne, cmap = plt.cm.gray)
    plt.axis("off")
    
    ax = figure.add_subplot(1, 2, 2)
    plt.imshow(imageTwo, cmap = plt.cm.gray)
    plt.axis("off")
    
    plt.show()
    
firstImage = cv2.imread("Image_1.png")
secondImage = cv2.imread("Image_2.png")

firstImage = cv2.cvtColor(firstImage, cv2.COLOR_BGR2GRAY)
secondImage = cv2.cvtColor(secondImage, cv2.COLOR_BGR2GRAY)

figure = plt.figure("Images")
images = ("Image 1", firstImage), ("Image 2", secondImage)

for (i, (name, image)) in enumerate(images):
    ax = figure.add_subplot (1,3, i + 1)
    ax.set_title(name)
    plt.imshow(image, cmap = plt.cm.gray)
    plt.axis("off")
    
plt.show()

imageComparison(firstImage, secondImage)
```
